## 6. Detailed Data Analysis

### 6.1 Generate Summary Report

```python
# Generate summary report
print("ðŸ“‹ Generating detailed summary report...")
summary_df = analyzer.generate_summary_report()
```

### 6.2 Key Indicator Comparison Analysis

```python
# Create key indicator comparison chart
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Indicator list
metrics = ['Shannon_Change', 'Capacity_Change', 'Individuals_Change', 
          'Species_Change', 'Disturbance_Change', 'Final_Corp_Capital']

metric_titles = [
    'Shannon Diversity Change', 'Carrying Capacity Change', 'Individual Count Change',
    'Species Count Change', 'Disturbance Level Change', 'Final Corporate Capital'
]

colors = ['skyblue', 'lightcoral', 'lightgreen', 'orange', 'plum']

for i, (metric, title) in enumerate(zip(metrics, metric_titles)):
    row = i // 3
    col = i % 3
    
    bars = axes[row, col].bar(summary_df['Scenario'], summary_df[metric], 
                             color=colors, alpha=0.8, edgecolor='black')
    axes[row, col].set_title(title, fontweight='bold', fontsize=12)
    axes[row, col].set_ylabel('Change Value' if 'Change' in metric else 'Value')
    axes[row, col].tick_params(axis='x', rotation=45)
    axes[row, col].grid(True, alpha=0.3)
    
    # Add numeric labels
    for bar in bars:
        height = bar.get_height()
        axes[row, col].text(bar.get_x() + bar.get_width()/2., height,
                           f'{height:.2f}', ha='center', va='bottom')

plt.suptitle('Key Indicator Comparison Analysis', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("ðŸ“Š Key indicator comparison chart generated")
```

### 6.3 Correlation Analysis

```python
# Calculate correlations between indicators
correlation_data = summary_df.select_dtypes(include=[np.number])
correlation_matrix = correlation_data.corr()

# Plot correlation heatmap
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='RdYlBu_r', 
            center=0, square=True, fmt='.3f', cbar_kws={'shrink': 0.8})
plt.title('Ecological Indicator Correlation Analysis', fontweight='bold', fontsize=14)
plt.tight_layout()
plt.show()

print("ðŸ”— Indicator correlation analysis completed")
```

## 7. Specific Scenario In-depth Analysis

### 7.1 Best vs Worst Scenario Comparison

```python
# Find scenarios with best and worst Shannon diversity performance
best_scenario = summary_df.loc[summary_df['Shannon_Change'].idxmax(), 'Scenario']
worst_scenario = summary_df.loc[summary_df['Shannon_Change'].idxmin(), 'Scenario']

print(f"ðŸŒŸ Best Shannon diversity scenario: {best_scenario}")
print(f"ðŸ’€ Worst Shannon diversity scenario: {worst_scenario}")

# Compare detailed performance of these two scenarios
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

scenarios_to_compare = [best_scenario.lower(), worst_scenario.lower()]
scenario_colors = ['green', 'red']
scenario_labels = [f'{best_scenario} (Best)', f'{worst_scenario} (Worst)']

# Historical comparison
metrics_to_plot = ['mean_shannon', 'mean_carrying_capacity', 'mean_individuals', 'total_individuals']
metric_titles = ['Shannon Diversity Index', 'Average Carrying Capacity', 'Average Individuals', 'Total Individuals']

for i, (metric, title) in enumerate(zip(metrics_to_plot, metric_titles)):
    row = i // 2
    col = i % 2
    
    for j, (scenario, color, label) in enumerate(zip(scenarios_to_compare, scenario_colors, scenario_labels)):
        if scenario in results:
            history = results[scenario]['history']
            steps = len(history[metric])
            axes[row, col].plot(range(steps), history[metric], 
                              color=color, linewidth=3, alpha=0.8, label=label)
    
    axes[row, col].set_title(title, fontweight='bold')
    axes[row, col].set_xlabel('Steps')
    axes[row, col].set_ylabel(title)
    axes[row, col].legend()
    axes[row, col].grid(True, alpha=0.3)

plt.suptitle('Best vs Worst Scenario Detailed Comparison', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("ðŸ“ˆ Best vs worst scenario comparison analysis completed")
```

### 7.2 Scenario Performance Ranking

```python
# Create comprehensive ranking
ranking_metrics = ['Shannon_Change', 'Individuals_Change', 'Final_Corp_Biodiv']
ranking_weights = [0.4, 0.3, 0.3]  # Weights: diversity 40%, individuals 30%, corporate biodiversity 30%

# Normalize scores
normalized_scores = summary_df[ranking_metrics].copy()
for metric in ranking_metrics:
    normalized_scores[metric] = (summary_df[metric] - summary_df[metric].min()) / (summary_df[metric].max() - summary_df[metric].min())

# Calculate comprehensive score
summary_df['Comprehensive_Score'] = sum(normalized_scores[metric] * weight for metric, weight in zip(ranking_metrics, ranking_weights))
summary_df_sorted = summary_df.sort_values('Comprehensive_Score', ascending=False)

# Visualize ranking
plt.figure(figsize=(12, 8))
colors = plt.cm.RdYlGn(np.linspace(0.3, 0.9, len(summary_df_sorted)))
bars = plt.bar(range(len(summary_df_sorted)), summary_df_sorted['Comprehensive_Score'], 
               color=colors, alpha=0.8, edgecolor='black', linewidth=2)

plt.xlabel('Scenario Ranking', fontweight='bold')
plt.ylabel('Comprehensive Score', fontweight='bold')
plt.title('Scenario Comprehensive Performance Ranking\n(Based on Shannon Diversity Change, Individual Count Change, Corporate Biodiversity Score)', 
          fontweight='bold', fontsize=14)
plt.xticks(range(len(summary_df_sorted)), summary_df_sorted['Scenario'], rotation=45)
plt.grid(True, alpha=0.3, axis='y')

# Add numeric labels and rankings
for i, (bar, score) in enumerate(zip(bars, summary_df_sorted['Comprehensive_Score'])):
    plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.01,
             f'#{i+1}\n{score:.3f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("ðŸ† Scenario comprehensive ranking:")
for i, row in summary_df_sorted.iterrows():
    print(f"   {i+1}. {row['Scenario']}: {row['Comprehensive_Score']:.3f}")
```
