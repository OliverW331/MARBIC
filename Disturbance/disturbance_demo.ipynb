# Tune the parameters of disturbance

```python
from disturbance import *
```

```python
# create a dummy effect_of_exploitation_mat for testing with all 1s with size n_grids x n_grids
n_grids = 50
effect_of_exploitation_mat = np.ones((n_grids, n_grids))
```

```python
#create disturbance object
disturbance = Disturbance(effect_of_exploitation_mat, alpha=0.1, beta=0.05, gamma=0.02,
                 k=1.0, f0=10.0, D_mode="exponential", B_mode="exponential", K_mode="exponential")
```

```python
disturbance.plot_disturbance_effects(max_f=100)
```
